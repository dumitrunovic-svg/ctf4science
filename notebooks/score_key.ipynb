```python
import os
from pathlib import Path

from ctf4science.data_module import _load_mat_file, _load_npz_file
from ctf4science.eval_module import evaluate_custom

repo_dir = Path(os.getcwd()).parent
dataset = "seismo"
```

```python
print("Evaluating on true results:")
if dataset in ["ODE_Lorenz", "PDE_KS"]:
    truth = _load_mat_file(repo_dir / "data" / dataset / "test" / "X1test.mat")
    prediction = _load_mat_file(repo_dir / "data" / dataset / "test" / "X1test.mat")
elif dataset == "seismo":
    truth = _load_npz_file(repo_dir / "data" / "seismo" / "test" / "X1test.npz")
    prediction = _load_npz_file(repo_dir / "data" / "seismo" / "test" / "X1test.npz")
else:
    raise ValueError(f"Unknown dataset: {dataset}")

for pair_id in range(1, 10):
    res_dict = evaluate_custom("ODE_Lorenz", pair_id, truth, prediction)
    print(f"pair_id {pair_id}")
    if "short_time" in res_dict:
        print(f"> stf: {res_dict['short_time']}")
    if "long_time" in res_dict:
        print(f"> ltf: {res_dict['long_time']}")
    if "reconstruction" in res_dict:
        print(f"> rec: {res_dict['reconstruction']}")
```

```python
print("Evaluating on zero as results:")
if dataset in ["ODE_Lorenz", "PDE_KS"]:
    truth = _load_mat_file(repo_dir / "data" / dataset / "test" / "X1test.mat")
    prediction = _load_mat_file(repo_dir / "data" / dataset / "test" / "X1test.mat")
elif dataset == "seismo":
    truth = _load_npz_file(repo_dir / "data" / "seismo" / "test" / "X1test.npz")
    prediction = _load_npz_file(repo_dir / "data" / "seismo" / "test" / "X1test.npz")
else:
    raise ValueError(f"Unknown dataset: {dataset}")

prediction = prediction * 0.0

for pair_id in range(1, 10):
    res_dict = evaluate_custom("ODE_Lorenz", pair_id, truth, prediction)
    print(f"pair_id {pair_id}")
    if "short_time" in res_dict:
        print(f"> stf: {res_dict['short_time']}")
    if "long_time" in res_dict:
        print(f"> ltf: {res_dict['long_time']}")
    if "reconstruction" in res_dict:
        print(f"> rec: {res_dict['reconstruction']}")
```
