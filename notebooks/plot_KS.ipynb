## Load Pickle

```python
if 0:
    fname = 'PDE_KS_p8_v0_r20.pkl'
    with open(Path(os.getcwd()).parent / 'models' / 'chronos-forecasting' / 'pickles' / fname, 'rb') as f:
        pred_data = pickle.load(f)


    print(pred_data.shape)
```

## Make plot

```python

test_data = _load_test_data("PDE_KS", 8)
pred_data = np.tile(test_data[-1,:], (1000,1)) # Just predict last time-step

viz = Visualization()
fig = viz.compare_prediction(test_data, [pred_data],
                                cbar_options={'ticks': 5, 'label': "PDE Value"})
fig.show()
```
