```python
import os
import yaml
import numpy as np
from pathlib import Path
from scipy.io import loadmat

lorenz_path = Path(os.getcwd()).parent / "data" / "ODE_Lorenz"
sst_path = Path(os.getcwd()).parent / "data" / "SST"
ocean_das_path = Path(os.getcwd()).parent / "data" / "ocean_das"
```

```python


def print_file_shapes(data_dir, shapes):
    files = sorted([f for f in Path(data_dir).rglob('*') if f.is_file()], key=lambda x: (x.stem, x.suffix))
    for filepath in files:
        file = filepath.name
        # Sort the file type by filename and then extension
        if file.endswith('.npy') or file.endswith('.mat') or file.endswith('.npz'):
            arr_shape = None
            if file.endswith('.npy'):
                arr = np.load(filepath)
                arr_shape = arr.shape
            elif file.endswith('.mat'):
                mat = loadmat(filepath)
                arr_candidates = [v for k, v in mat.items() if not k.startswith('__') and isinstance(v, np.ndarray)]
                if arr_candidates:
                    arr = arr_candidates[0]
                    arr_shape = arr.shape
                else:
                    print(f"{file}: No array found in .mat file")
                    continue  # Skip to next file
            elif file.endswith('.npz'):
                npz_data = np.load(filepath)
                keys = list(npz_data.keys())
                arr = npz_data[keys[0]]
                arr_shape = arr.shape
            # Calculate min and max values
            arr_min = arr.min()
            arr_max = arr.max()
            # Check against expected shape from shapes dict, if available
            expected_shape = shapes.get(file, None)
            if expected_shape is not None:
                matches = tuple(expected_shape) == tuple(arr_shape)
                status = "OK" if matches else f"Mismatch (expected {tuple(expected_shape)})"
                print(f"{file}: {arr_shape} [min={arr_min:.6f}, max={arr_max:.6f}] [{status}]")
            else:
                print(f"{file}: {arr_shape} [min={arr_min:.6f}, max={arr_max:.6f}] [No expected shape found]")
        else:
            print(f"{file}: Unsupported file type")


# Example usage:
shapes = yaml.safe_load(open(ocean_das_path / "ocean_das.yaml"))['metadata']['matrix_shapes']
print("ocean_das:")
print_file_shapes(ocean_das_path, shapes)

shapes = yaml.safe_load(open(lorenz_path / "ODE_Lorenz.yaml"))['metadata']['matrix_shapes']
print("lorenz:")
print_file_shapes(lorenz_path, shapes)
```
